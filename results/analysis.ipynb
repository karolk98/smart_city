```python
def roundp(y, p=0):
    res = y.copy()
    for i in range(p):
        res += y.shift(periods=i)
        res += y.shift(periods=-i)
    return res/(2*p+1)

def plot2(name ):
    df = pd.read_csv(f'dump2/{name}.csv')
    x = df['learning evaluation instances'].astype(int)
    y = df['classifications correct (percent)']
    y = roundp(y,10)
    plt.plot(x,y, label=name)


# plot2('seq-marline-srp')
# plot2('seq-marline-boost')
plot2('seq-melanie-srp')
plot2('tar-melanie-srp')
plot2('tar-srp')
plt.xlabel('Instances')
plt.ylabel('Accuracy[%]')
plt.legend()
# plt.ylim((60,100))
```

```python
from os import walk
import statistics as stats

def mean_acc(path):
    f = []
    for (dirpath, dirnames, filenames) in walk(path):
        f.extend(filenames)
    f = sorted(f)
    for file in f:
        print(file)
        print(stats.mean(pd.read_csv(f'{path}/{file}')['classifications correct (percent)']))

def mean_kapp(path):
    f = []
    for (dirpath, dirnames, filenames) in walk(path):
        f.extend(filenames)
    f = sorted(f)
    for file in f:
        print(file)
        print(stats.mean(pd.read_csv(f'{path}/{file}')['Kappa Statistic (percent)'].dropna()))

mean_acc('dump2')
```

```python
def latex(path, datasets, models, measure = 'acc'):
    df = pd.DataFrame()
    df['Model'] = models.values()
    if measure == 'acc':
        measure = 'classifications correct (percent)'
    else:
        measure = 'Kappa Statistic (percent)'
    for data in datasets.keys():
        l = []
        for model in models.keys():
            file = path + '/'+data+'_'+model+'.csv'
            col = pd.read_csv(file)[measure]
            col = col[(col != '?')] 
            col = pd.to_numeric(col, downcast="float").dropna()
            # col = col[~(col <= 1)] 
            l.append("%.2f" % stats.mean(col))
        df[datasets[data]] = l
    return df

models = {
    'maj': 'Majority',
    # 'arf': 'ARF',
    'srp': 'SRP',
    # 'melanie-boost': 'Melanie Boost',
    # 'melanie': 'Melanie SRP',
    # 'smelanie': 'Smelanie SRP',
    # 'melanie2': 'Melanie+ SRP',
    # 'MARLINE': 'MARLINE SRP',
    # 'blast': 'Blast'
}

datasets = {
    # 'hist-tar': 'HistTar',
    # 'alt': 'Alt',
    # 'hist-alt': 'HistAlt',
    # 'seq': 'Seq',
    # 'hist-seq': 'HistSeq',
    # 'seq-ind': 'SeqInd',
    # 'hist-seq-ind': 'HistSeqInd',
    # 'long-hist-seq': 'LongHistSeq',
    # 'abr': 'Abr',
    'tar': 'Tar',
    'hist-tar': 'HistTar',
    'tar-ind': 'TarInd',
    'hist-tar-ind': 'HistTarInd'
}

df = latex('dump10', models=models, datasets=datasets,measure='kapp')
print(df.to_latex(index=False).replace('\\\\\n','\\\\ \\hline\n'))
```
